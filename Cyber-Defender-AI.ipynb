```python
import random

# Define the game assets and AI tools
assets = {
    "Employee Data": 100,
    "Financial Information": 100,
    "Intellectual Property": 100
}

ai_tools = {
    "AI Email Filter": 0.8,  # 80% success rate
    "AI Antivirus": 0.7,     # 70% success rate
    "AI Network Monitor": 0.9  # 90% success rate
}

# Define possible threats
threats = ["Phishing Email", "Malware Attack", "Data Breach"]

# Initialize score
score = 0

# Game loop
for day in range(1, 11):  # Simulate 10 days
    print(f"Day {day}:")
    threat = random.choice(threats)
    print(f"A new threat has emerged: {threat}")

    # Display AI tools
    print("Available AI Tools:")
    for i, tool in enumerate(ai_tools):
        print(f"{i+1}. {tool} (Success Rate: {int(ai_tools[tool]*100)}%)")

    # Get a valid player choice
    while True:
        try:
            choice = int(input("Choose a tool to defend against the threat (1/2/3): ")) - 1
            if choice in range(len(ai_tools)):
                break
            else:
                print("Invalid choice. Please select a number between 1 and 3.")
        except ValueError:
            print("Invalid input. Please enter a number.")

    chosen_tool = list(ai_tools.keys())[choice]

    # Determine if the defense is successful
    if random.random() <= ai_tools[chosen_tool]:
        print(f"Success! The {threat} was neutralized using {chosen_tool}.")
        score += 10
    else:
        print(f"Failure! The {threat} breached the defenses.")
        asset_damaged = random.choice(list(assets.keys()))
        assets[asset_damaged] -= random.randint(10, 30)
        score -= 5

    # Display the current status of assets
    print("Current Asset Status:")
    for asset, value in assets.items():
        print(f"{asset}: {value}%")

    print("\n")

# Game Over
print("Game Over!")
print(f"Final Score: {score}")
print("Final Asset Status:")
for asset, value in assets.items():
    print(f"{asset}: {value}%")

if score > 50:
    print("You did a great job protecting the company!")
else:
    print("The company suffered significant losses. Better luck next time!")
```
